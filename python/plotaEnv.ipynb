```python
import serial
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', 10)
```

```python
def formata_hora(s):
    dt = datetime.fromisoformat(s)
    return dt.strftime("%H:%M:%S")
```

```python
df=pd.read_csv("saida_serial.csv",names=["tempo","countTHENV","valEnv"])
df
```

```python
df1 = df.copy()
df1.tempo = df.tempo.apply(formata_hora)
df1
```

```python
plt.figure(figsize=(18,6))
plt.plot(df1["tempo"],df1["valEnv"])
plt.xticks(df1["tempo"][::len(df1)//30],rotation=90)
plt.ylim(8,50)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.plot(df1["tempo"],df1["countTHENV"])
plt.xticks(df1["tempo"][::len(df1)//30],rotation=90)
plt.show()
```

```python
def plota_ultimos(n):
    df=pd.read_csv("saida_serial.csv",names=["tempo","countTHENV","valEnv"])
    df1 = df[-n:].copy()
    df1.tempo = df.tempo.apply(formata_hora)

    plt.figure(figsize=(18,6))
    plt.plot(df1["tempo"],df1["valEnv"])
    plt.xticks(df1["tempo"][::len(df1)//30],rotation=90)
    plt.ylim(8,50)
    plt.show()

    plt.figure(figsize=(18,6))
    plt.plot(df1["tempo"],df1["countTHENV"])
    plt.xticks(df1["tempo"][::len(df1)//30],rotation=90)
    plt.show()
```

```python
plota_ultimos(10000)
```
